# Отток клиентов из банка

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Нам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Необходимо построить модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59.

Дополнительно необходимо измерять *AUC-ROC*, сравнивая её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

**Описание данных**

Каждая строчка в датасете является информацией о конкретном клиенте банка. В базе данных присутствуют следующие столбцы, которые будут выступать **признаками** в дальнейшем анализе:

Например, врятли от фамилии человека или от его уникального идентификатора может зависеть его доход (даже если это так, врятли это будет применимо к выборке с другими клиентами банка)

Разделим данные на тренировочную, валидационную и тестовую выборки

```python
features = data_adjusted.drop('Exited', axis=1)
target = data_adjusted['Exited']

# Разделим выборку на тренировочную и вылидационную
features_train, features_valid, target_train, target_valid = (
    train_test_split(features, target, test_size=0.4, random_state=12345, stratify=target)
)

# Разделим валидационную выборку на валидационную и тестовую
features_valid, features_test, target_valid, target_test = (
    train_test_split(features_valid, target_valid, test_size=0.5, random_state=12345,
                     stratify=target_valid)
)

#features_train.shape
#features_valid.shape
#features_test.shape
#target_train.shape
#target_valid.shape
#target_test.shape
```

Также, для подготовки признаков необходимо провести масштабирование признаков, так как разброс и диапазон данных в столбцах `Balance`, `EstimatedSalary` очень сильно отличаются от остальных данных, например от `Age`. Если не привести данные к одному масштабу, при обучении моделей это может привести к восприятию алгоритмом признака с большим разбросом и диапазоном как наиболее важного, что может привести к переобучению модели

```python
# Запишем все количественные признаки в список numeric
numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

# Используем метод StandardScaler
# Он приводит все значения к стандартному нормальному распределению
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

#features_train.head()
#features_valid.head()
#features_test.head()
```

Для подготовки признаков класса object к анализу и построению модели, используем технику OHE. Кодирование признаков необходимо, так как в датасете присутствуют категориальные признаки: `Geography`, `Gender`. Техника порядкового кодирования не подойдет, так как в качестве одна из рассматриваемых моделей - логистическая регрессия, для которой данная техника кодирования не подходит.  

```python
# Сохраним категориальные признаки в список ohe_features
ohe_features = features_train.select_dtypes(include='object').columns.to_list()

# Создадим encoder и обучим его на тренировочных данных
encoder = OneHotEncoder(handle_unknown='ignore', drop='first', sparse=False)
encoder.fit(features_train[ohe_features])

features_train[encoder.get_feature_names_out()] = encoder.transform(features_train[ohe_features])
features_train = features_train.drop(ohe_features, axis=1)
#features_train.head()

features_valid[encoder.get_feature_names_out()] = encoder.transform(features_valid[ohe_features])
features_valid = features_valid.drop(ohe_features, axis=1)
#features_valid.head()

features_test[encoder.get_feature_names_out()] = encoder.transform(features_test[ohe_features])
features_test = features_test.drop(ohe_features, axis=1)
#features_test.head()
```

### Построение моделей без учета дисбаланса

Для оптимизации кода напишем функции, которые принимают на вход тренировочные и валидационные выборки, и параметр class_weight. Функции будут подбирать оптимальные гипермараметры для моделей, основываясь на показателе f1. Функции будут возвращать оптимальные модели, подобранные гиперпараметры и знечние мерки f1. Также, функции будут выводить полученные параметры, значение f1, матрицу ошибок и график AUC-ROC на экран

```python
# Функция для регрессии
def regression_builder(features_train, target_train, features_valid, target_valid, class_weight=None):
    model = LogisticRegression(random_state=12345, max_iter=1000, solver='liblinear', 
                               class_weight=class_weight)
    model.fit(features_train, target_train)
    predictions_valid = model.predict(features_valid)
    f1 = f1_score(target_valid, predictions_valid)

    print('Матрица ошибок')
    display(pd.DataFrame(confusion_matrix(target_valid, predictions_valid)))
    
    print(f'Значение f1-score на валидационной выборке: {f1}')

    probabilities_one = model.predict_proba(features_valid)[:, 1]
    auc_score = roc_auc_score(target_valid, probabilities_one)
    print(f'\nЗначение метрики roc-auc score на валидационной выборке: {auc_score}')
    fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one)

    plt.figure()
    plt.plot(fpr, tpr)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая линейной регрессии')
    plt.show() 
    
    return model, f1
```

```python
# Функция для дерева решений
def tree_builder(features_train, target_train, features_valid, target_valid, class_weight=None):
    best_model = None
    best_depth = 0
    best_f1 = 0

    for depth in range(1, 11):
    
        model = DecisionTreeClassifier(max_depth=depth, random_state=12345, class_weight=class_weight)
        model.fit(features_train, target_train)
    
        predictions_valid = model.predict(features_valid)
        f1 = f1_score(target_valid, predictions_valid)
    
        if f1 > best_f1:
            best_model = model
            best_depth = depth
            best_f1 = f1
        
    print(f'Лучший результат модели дерева решений:\nГлубина - {best_depth}\n'
          f'F1 - {best_f1}')

    best_predictions_valid = best_model.predict(features_valid)

    print('\nМатрица ошибок')
    display(pd.DataFrame(confusion_matrix(target_valid, best_predictions_valid)))

    probabilities_one = best_model.predict_proba(features_valid)[:, 1]
    auc_score = roc_auc_score(target_valid, probabilities_one)
    print(f'\nЗначение метрики roc-auc score на валидационной выборке: {auc_score}')
    fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one)

    plt.figure()
    plt.plot(fpr, tpr)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая дерева решений')
    plt.show()
    
    return best_model, best_depth, best_f1
```

```python
# Функция для случайного леса
def forest_builder(features_train, target_train, features_valid, target_valid, class_weight=None):
    best_model = None
    best_f1 = 0
    best_depth = 0
    best_est = 0

    for est in range(10, 51, 10):
        for depth in range(1, 20):
            model = RandomForestClassifier(n_estimators=est, max_depth=depth, random_state=12345, 
                                          class_weight=class_weight)
            model.fit(features_train, target_train)
        
            predictions_valid = model.predict(features_valid)
            f1 = f1_score(target_valid, predictions_valid)
        
            if f1 > best_f1:
                best_model = model
                best_depth = depth
                best_est = est
                best_f1 = f1
            
    print(f'Лучший результат модели случайного леса:\nГлубина - {best_depth}\n'
      f'Число деревьев в случайном лесу - {best_est}\nF1 - {best_f1}')

    best_predictions_valid = best_model.predict(features_valid)

    print('\nМатрица ошибок')
    display(pd.DataFrame(confusion_matrix(target_valid, best_predictions_valid)))

    probabilities_one = best_model.predict_proba(features_valid)[:, 1]
    auc_score = roc_auc_score(target_valid, probabilities_one)
    print(f'\nЗначение метрики roc-auc score на валидационной выборке: {auc_score}')
    fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one)

    plt.figure()
    plt.plot(fpr, tpr)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая случайного леса')
    plt.show() 
    
    return best_model, best_depth, best_est, best_f1
```

#### Дерево решений

```python
model_tree, best_depth_tree, f1_tree = (
    tree_builder(features_train, target_train, features_valid, target_valid)
)
```

#### Случайный лес

```python
model_forest, best_depth_forest, best_est, f1_forest = (
    forest_builder(features_train, target_train, features_valid, target_valid)
)
```

Таким образом, без борьбы с дисбалансом **случайный лес** дал наилучший результат показателя f1 на валидационной выборке - примерно 0.61.

### Upsampling

```python
# Напишем функцию upsample:
# Ее аргументами являются признаки, целевой признак и на сколько необходимо увеличить выборку
# Функция возращает увеличенную в размере выборку

def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

```python
# Посчитаем параметр round как отношение размера большего класса выборки к меньшему

repeat = round(len(data[data['Exited'] == 0]) / len(data[data['Exited'] == 1]))
# repeat
```

```python
# Сохраним выборки после использования функции unsample()

features_upsampled, target_upsampled = upsample(features_train, target_train, repeat)
#len(features_upsampled) / len(target_upsampled)
```

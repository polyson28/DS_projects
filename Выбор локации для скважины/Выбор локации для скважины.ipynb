Мы работаем в добывающей компании «ГлавРосГосНефть». Нужно решить, где бурить новую скважину.

Нам предоставлены пробы нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов. Необходимо построить модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Анализ возможной прибыли и рисков нужно провести техникой *Bootstrap. Необходимо оставить только те регионы, в которых вероятность убытков не более 2.5%*

Шаги для выбора локации:

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

Нам доступны данные по геологической разведке трех регионов со следующими признаками:  
`id` - уникальный идентификатор скажины  
`f0`, `f1`, `f2` - три признака точек  
Целевой признак:  
`product` - объем запасов в скважине (тыс. баррелей)

## Обучение и проверка модели

```python
def model_constructor(features_train, target_train, features_valid, target_valid):
    # Построим модель линейной регрессии и обучим ее на тренировочной выборке
    model = LinearRegression()
    model.fit(features_train[numeric], target_train)

    # Построим предсказания модели на валидационной выборке
    predicted_valid = model.predict(features_valid[numeric])

    # Посчитаем среднеквадратичную ошибку модели и оценим средний запас предсказанного сырья
    mean_predicted = predicted_valid.mean()
    rmse = mean_squared_error(target_valid, predicted_valid, squared=False)

    print(f'Средний запас сырья, предсказанный моделью: {mean_predicted}\n'
          f'RMSE модели: {rmse}')
    
    return model, predicted_valid, mean_predicted, rmse
```

Посчитаем объем сырья, при котором точка будет безубыточна. Для этого сохраним все необходимые значения: бюджета для разработки новых скважин в регионе, прибыль с каждого барреля и общее количество скважин для разработкм

```python
BUDGET = 10000000000 # общий бюджет
PROFIT_PER_UNIT = 450000 # прибыль с одного барреля
WELL_NUMBER = 200 # количество скважин в каждом регионе
```

## Расчёт прибыли и рисков 

### Оценка прибыли

```python
# Сохраним константы, необходимые для расчетов
BEST_WELL = 200
TOTAL_WELL = 500
```

```python
# Напишем функцию расчета прибыли
def profit(data):
    
    # Отсортируем точки по прибыльности предсказанного уровня сырья
    data_sorted = data.sort_values(by='predicted', ascending=False)
    
    # Сохраним реальный уровень сырья по 200 точкам с наибольшим предсказанным уровнем
    # и посчитаем прибыль
    target = data_sorted['target'][:BEST_WELL]
    profit = sum(target) * PROFIT_PER_UNIT - BUDGET
    
    return profit
```

```python
# Напишем функцию бутсрапа
def bootstrap(data, iterations=1000, region=None):
    
    # Напишем цикл с 1000 итераций
    # В цикле строится sample из данных, убираются дубликаты по столбцу 'id'
    # Для расчета прибыли используется ранее написанная функция profit
    state = np.random.RandomState(12345)

    values = []
    for i in range(iterations):
        sample = data.sample(n=TOTAL_WELL, replace=True, random_state=state)
        profit_calculated = profit(sample)
        values.append(profit_calculated)
        
    values = pd.Series(values)
    
    # Выведем параметры распределения
    description = pd.DataFrame(values.describe())
    description.columns = [f'Параметры распределения в регионе {region}']
    display(description)
    
    mean = values.mean()
    print(f'Оцененная средняя прибыль для региона {region}: {mean}')
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    
    # Построим график плотности (density function)
    sns.kdeplot(values)
    sns.set(rc={'figure.figsize':(8, 5)})
    plt.title(f'График распределения прибыли в регионе {region}')
    plt.xlabel('Прибыль')
    plt.ylabel('Плотность')
    plt.axvline(x=mean, color='r', label='среднее')
    plt.axvline(x=lower, color='g', ls='--', label='доверитеьльный интервал')
    plt.legend()
    plt.axvline(x=upper, color='g', ls='--')
    plt.show()
    
    return values
```

### Оценка рисков

Для того, чтобы оценить риски убытков, посчитаем 95% доверительный интервал для средней прибыли и вероятность того, что средняя прибыль окажется отрицательной для каждого региона

```python
value_list = [values_1, values_2, values_3]

for i in range(3):
    print(f'Регион {i+1}')
    lower = value_list[i].quantile(0.025)
    upper = value_list[i].quantile(0.975)
    print(f'Доверительный интервал для средней прибыли: ({lower}, {upper})')
    risk = (value_list[i] < 0).mean()
    print(f'Риск убыточности: {risk:.1%}\n')
```

Таким образом, наименьший риск убыточности во втором регионе - примерно 2.3%, что также является допустимым уровнем риска, так как полученное значение меньше 2.5%. Остальные регионы не подходят по допутимому уровню риска.   
  
Также, оценивая доверительные интервалы, в первом и третьем регионах в 95% доверительный интервал для средней прибыли входят отрицательные значения, в отличие от второго региона, что позволяет сделать вывод о том, что второй регион наиболее вероятно принесет положительную прибыль

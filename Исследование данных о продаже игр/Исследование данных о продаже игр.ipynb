### Изменение типов данных

```python
data.query('user_score == "tbd"')['year_of_release'].unique()
```

*Значения 'tbd' есть в разные года, не только в наиболее близкие к 2016. Так как это неопределенная оценка пользователей, ее можно расценивать как пропуски. При анализе отсутствующих значений в дальнейшей предобработке данных нужно будет проанализировать причину появления пропусков в данном столбце, и определить, возможно ли их заполнить каким-то значением* 

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')
# data['user_score'].unique()
```

*Так как пропуски составляют незначительную часть датасета, удалим строчки с пропущенными значениями в данном столбце*

### Обработка пропусков

```python
data.isna().sum()
```

*Пропуски в столбцах `name` и `genre` малочисленны, так что можно удалить такие ячейки*

```python
data = data.dropna(subset=['name', 'genre'])
```

*Рассмотрим пропуски в столбцах `critic_score` и `user_score`. В данных колонках пропущенные значения составляют довольно большую часть, относительно всего датасета, а именно:*

```python
critic_na_percentage = data['critic_score'].isna().sum() / len(data)
user_na_percentage = data['user_score'].isna().sum() / len(data)

print(f'Оставшиеся пропуски в столбце с оценками критиков составляют {(critic_na_percentage):.1%}'
      f' от всего датасета')
print(f'Оставшиеся пропуски в столбце с оценками пользователей составляют '
      f'{(user_na_percentage):.1%} от всего датасета')
```

*Можно предположить, что оценки пользователей и критиков не выставлены только у игр, выпустившихся недавно. Построим гистаграмму количества пропущенных значений в столбцах `user_score` и `critic_score` по годам выпуска игр, чтобы проверить наше предположение.*

```python
# Создадим вспомогательную таблицу na_scores с количеством пропущенных значений в указанных столбцах

na_critic_scores = data[data['critic_score'].isna()]
na_user_scores = data[data['user_score'].isna()]

na_critic_scores_pivot = na_critic_scores.pivot_table(index=['year_of_release'], values=['name'],
                                                      aggfunc='count')
na_critic_scores_pivot.columns = ['critic_score_na_count']

na_user_scores_pivot = na_user_scores.pivot_table(index=['year_of_release'], values=['name'],
                                                  aggfunc='count')
na_user_scores_pivot.columns = ['user_score_na_count']

na_scores = na_critic_scores_pivot.merge(na_user_scores_pivot, on='year_of_release', how='outer')
# na_scores
```

```python
# Построим график зависимости наличия пропуска в столбце с оценками игр от года выпуска
# Чтобы проверить предположение, что большинство пропусков составляют новые игры

r1 = np.arange(len(na_scores))
r2 = [x + 0.4 for x in r1]

plt.figure(figsize=(10, 7))
plt.bar(r1, na_scores['critic_score_na_count'], width=0.4, label='пропуски в оценках критиков')
plt.bar(r2, na_scores['user_score_na_count'], width=0.4, 
        label='пропуски в пользовательских оценках')

ticks = [0, 5, 10, 15, 20, 25, 30, 35]
labels = []
for i in range(0, len(ticks)):
    labels.append(min(data['year_of_release']) + 5*i)

plt.xlabel('года')
plt.ylabel('Количество пропусков')
plt.title('Распределение пропусков по годам')
plt.xticks(ticks=ticks, labels=labels)

plt.legend()
plt.show()
```

*На графике количества пропусков в столбцах `user_score` и `critic_score` прослеживается 2 пика: 1998 и 2009 года, что противоречит нашему изначальному предположению о том, что значения пропущены в основном для недавно выпустившихся игр. Также, можно заметить, что два столбца распределены примерно одинаково по годам, что позволяет сделать предположение о том, что в строчках с пропусками в столбце `user_score`, также будет пропуск в столбце `critic_score`. Проверим наше предположение*

```python
both_user_critic_na = len(data[(data['user_score'].isna()) & (data['critic_score'].isna())])
user_na = data['user_score'].isna().sum()
critic_na = data['critic_score'].isna().sum()

print(f'Количество строк, где пропущены оценки и пользователей, и критиков: {both_user_critic_na}')
print(f'Количество строк с пропущенными оценками пользователей: {user_na}')
print(f'Количество строк с пропущенными оценками критиков: {critic_na}')
```

*Наше предположение подтвердилось: в большинстве строчек с пропусками в одном столбце, отсутствует значение и в другом столбце. 
<br>Возможной причиной возникновения пропусков могла послужить непопулярность игр: не удалось собрать достаточное количество оценок. Однако, пропуски в указанных значениях составляют около половины датасета. Следовательно, нужно предупредить разработчиков, ответственных за выгрузку данных, о данной проблеме. Возможно, произошел сбой при выгрузке данных.
<br>Заполнение такого большого количества пропусков каким-либо значением (например, медианой) может привести к сильному искажению статистики. Следовательно, заполнить пропуски в оценках пользователей и критиков мы заполнить не можем.*

*Рассмотрим пропуски в столбце `rating`, значения которого - рейтинг от организации ESRB. Так как пропусков в данном столбце тоже довольно много, значения данного столбца нельзя заполнить каким-либо значением (например, модой), так как это может привести к неверным выводам в последующем анализе. 
<br> Попробуем установить причину пропусков. Можно предположить, что пропуски также, как и в случае оценок, возникли из-за недостаточного объема данных для присваивания рейтинга. Чтобы проверить предположение, проверим, сколько строчек, где пропущено значение рейтинга и оценок одновременно*

```python
scores_rating_na = len(data[(data['rating'].isna()) & (data['user_score'].isna())
                            & (data['critic_score'].isna())])
rating_na = len(data[data['rating'].isna()])

print(f'Количество строк, где пропущены оценки пользователей, критиков и рейтинг: '
      f'{scores_rating_na}')
print(f'Количество строк с пропущенным рейтингом: {rating_na}')
```

*Наше предположение подтвердилось, так как в датасете довольно большое количество строчек, где пропущены и оценки, и рейтинг игр. Значит, пропуски могли возикнуть из-за непопулярности игр (не хватило данных для оценки и выставления рейтинга). Однако, пропуски в указанных столбцах составляют довольно большую часть датасета, о чем нужно предупредить разработчиков.*

```python
deleted_percentage = (initial_dataset_size - len(data)) / initial_dataset_size
print(f'После предобработки данных было удалено {deleted_percentage:.2%} датасета')
```

## Портрет пользователя каждого региона

*Сравним распределение самых популярных платофрм, жанров игр для пользователей из разных стран, и проверим, как рейтинг ESRB влияет на продажи в отдельном регионе*

```python
# Напишем функцию, которая принимает критерий сравнения в качестве аргумента
# Функция возвращает три круговых диаграммы по регионам
# и таблицы с самыми популярными платформами / жанрами

def pie_by_region(criteria, number_of_shares):
    for region in ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']:
        
        pie_pivot = data_filtered.pivot_table(index=[criteria], values=region, aggfunc='sum')
        
        country = region.split('_')[0].upper()
        table_header = f'Продажи в {country} (в миллионах)'
        
        pie_pivot.columns = [table_header]
        pie_pivot = pie_pivot.sort_values(by=table_header, ascending=False).head(number_of_shares)
        pie_pivot['Доля на рынке'] = (pie_pivot[table_header] /
                                      sum(pie_pivot[table_header])).round(2)
        display(pie_pivot)
        
        pie_pivot[table_header].plot(kind='pie')
        plt.ylabel('')
        plt.title(f'Распределение продаж в {country}')
        plt.show()
```

```python
# Построим круговые диаграммы долей продаж на 5 самых популярных платформах по регионам

pie_by_region('platform', 5)
```

*Можно заметить схожесть в популярных платформах в разбивке по странам: 3DS, PS3 и PS4 входят в пятерку лидеров во всех регионах, X360 и XOne - во всех, кроме Японии.

В Японии среди самых популярных платформ присутствую PSV и WiiU, хотя в остальных странах данные платформы не такие популярные.*

```python
# Построим круговые диаграммы долей продаж для 5 самых популярных жанров по регионам

pie_by_region('genre', 5)
```

```python
# Разделим датасет на две выборки по платформам Xbox One и PC
# Проверим размеры получившихся выборок

sample_xbox = data_filtered.query('platform == "XOne"')['user_score'].dropna().reset_index(drop=True)
print(f'Размер выборки по платформе XOne: {len(sample_xbox)}')

sample_pc = data_filtered.query('platform == "PC"')['user_score'].dropna().reset_index(drop=True)
print(f'Размер выборки по платформе PC: {len(sample_pc)}')
```

Н_0: средние пользовательские рейтинги равны
<br>H_1: средние пользовательские рейтинги разные

Возьмем для анализа альфу равную 5%. Размеры выборок достаточные (больше 30), однако разные по значению. Так как размеры выборок отличаются и нет оснований полагать, что дисперсии выборок равны, определим параметр `equal_var=False`

```python
results_1 = st.ttest_ind(sample_xbox, sample_pc, equal_var=False)
alpha = .05

print('p-значение:', results_1.pvalue)

if results_1.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
